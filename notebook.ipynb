```python
# Tamanho da imagem
fig, ax = plt.subplots(figsize=(16,10))

plt.style.use('seaborn')

# Cores de fundo
backgorund = '#f5f5f5'
ax.set_facecolor(backgorund)
fig.set_facecolor(backgorund)

# Subtítulo da figura
plt.suptitle("Análise do tempo de Serviço \npelo salário", fontsize = 18, color='black', fontweight=400)

# Parâmetros para a configuração do report
linhas = 2
colunas = 2

plt.subplot(linhas, colunas, 1)
plt.title('Análise de Renda')
sns.kdeplot(db['Salary'], color = 'magenta');

plt.subplot(linhas, colunas, 2)
plt.title('Análise da Experiência')
sns.kdeplot(db['YearsExperience'], color = 'red')

plt.subplot(linhas, colunas, 3)
plt.title('Análise de Renda')
sns.boxplot(db['Salary'], color = 'magenta')

plt.subplot(linhas, colunas, 4)
plt.title('Análise da Experiência')
sns.boxplot(db['YearsExperience'], color = 'green')

# Ajsutando o Espaço
plt.subplots_adjust( hspace=0.35, wspace=0.15 )
```

```python
# Tamanho da imagem
fig, ax = plt.subplots(figsize=(16,10))

plt.style.use('seaborn')

# Cores de fundo
backgorund = '#f5f5f5'
ax.set_facecolor(backgorund)
fig.set_facecolor(backgorund)

# Subtítulo da figura
plt.suptitle("Análise do tempo de Serviço \npelo salário", fontsize = 18, color='black', fontweight=400)

# Parâmetros para a configuração do report
linhas = 2
colunas = 2 

plt.subplot(linhas, colunas, 1)
plt.title('Análise do Salário pelos anos de experiência')
sns.scatterplot(data=db, x='Salary', y='YearsExperience')

plt.subplot(linhas, colunas, 2)
plt.title('Análise do Salário pelos anos de experiência')
sns.regplot(data=db, x='Salary', y='YearsExperience')

plt.subplot(linhas, colunas, 3)
plt.title('Correlação de Pearson')
cor = db.corr('pearson')
sns.heatmap(cor, annot=True)

plt.subplot(linhas, colunas, 4)
plt.title('Correlação de Spearman')
cor2 = db.corr('spearman')
sns.heatmap(cor2, annot=True);

# Ajsutando o Espaço
plt.subplots_adjust( hspace=0.35, wspace=0.15 )
```
